## Download a dataset

```
#@title Download a dataset

def download_dataset(dataset_name):
    if dataset_name == "":
        return

    user_name, dataset_name = dataset_name.split("/")
    %cd {INPUT_FOLDER}
    !rm -rf {dataset_name} && echo "Removed existing version"
    !kaggle datasets download {user_name}/{dataset_name} -p {dataset_name} --unzip --force && echo "Downloaded"

dataset_name = "xhlulu/vinbigdata" #@param {type:"string"}
download_dataset(dataset_name)
```
